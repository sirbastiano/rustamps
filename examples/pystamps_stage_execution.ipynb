# Exploring stage execution with pySTAMPS

This notebook explains how pySTAMPS execution works without forcing a heavy processing run.

The most important beginner habit is this: **always preview with `--dry-run` first**.


## What a stage range means

The pySTAMPS workflow is divided into numbered stages.

In practice, that means you can ask pySTAMPS to:

- preview all stages
- run only an early subset
- resume at a later stage if earlier outputs already exist

This staged design is useful because radar processing can be expensive, and you often want to isolate one part of the workflow.


```python
from pathlib import Path

DATASET_MAIN = Path('inputs_and_outputs/InSAR_dataset_test')
DATASET_STAGE8 = Path('inputs_and_outputs/InSAR_dataset_test_stage8diag')

for path in (DATASET_MAIN, DATASET_STAGE8):
    print(f'{path}:', 'present' if path.exists() else 'missing')
```

## Dry-run first

A dry-run shows what pySTAMPS would try to do without performing the heavy work. This is the default educational mode for this notebook.


```python
dry_run_command = (
    'uv run pystamps run --dataset inputs_and_outputs/InSAR_dataset_test '
    '--start-step 1 --end-step 8 --dry-run'
)
print(dry_run_command)
```

## Narrow the stage range when learning

Beginners often benefit from understanding a smaller slice of the workflow first.

For example, you might preview just an early subset or focus on the later merged stages.


```python
stage_commands = [
    'uv run pystamps run --dataset inputs_and_outputs/InSAR_dataset_test --start-step 1 --end-step 3 --dry-run',
    'uv run pystamps run --dataset inputs_and_outputs/InSAR_dataset_test --start-step 6 --end-step 8 --dry-run',
]
for command in stage_commands:
    print(command)
```

## Real execution is optional and non-default

If you later decide to actually execute stages, do it on a copy of the dataset, not on your only original folder.

This notebook only prints the command below. It does not run it.


```python
real_run_command = (
    '# Optional non-default step\n'
    'cp -a inputs_and_outputs/InSAR_dataset_test inputs_and_outputs/InSAR_dataset_test_run_copy\n'
    'uv run pystamps run --dataset inputs_and_outputs/InSAR_dataset_test_run_copy --start-step 1 --end-step 3'
)
print(real_run_command)
```

## Why external tools matter

Some workflows depend on tools outside Python, especially `triangle` and `snaphu`.

That is another reason to prefer `--dry-run` while learning: it lets you understand the command shape before you debug environment dependencies.
