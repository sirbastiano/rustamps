## What pySTAMPS does

`pystamps` is a Python implementation of a staged processing workflow for radar interferometry datasets.

In plain language:

- A **dataset** is a folder containing files that describe a radar scene and intermediate processing outputs.
- A **stage** is one step in the workflow. pySTAMPS has stages 1 through 8.
- A **persistent scatterer** is a ground point that stays stable enough across many images to be useful for measurement.
- An **interferogram** is a comparison between two radar observations. It helps estimate how the phase changed.
- **Verification** means comparing pySTAMPS outputs with a known reference or golden dataset.

You do not need to master the mathematics before using the tools safely. The first practical skills are learning how to inspect a dataset, preview what pySTAMPS would do, and verify outputs carefully.

## The three commands you will use most often

pySTAMPS can do many things, but most day-to-day exploration starts with these commands:

- `pystamps status`: inspect dataset structure and inferred stage progress.
- `pystamps run`: execute a stage range, usually on a copy of a dataset.
- `pystamps verify`: compare one run folder against a golden reference folder.


```python
commands = [
    'uv run pystamps status --dataset inputs_and_outputs/InSAR_dataset_test',
    'uv run pystamps run --dataset inputs_and_outputs/InSAR_dataset_test --start-step 1 --end-step 8 --dry-run',
    'uv run pystamps verify --run inputs_and_outputs/InSAR_dataset_test --golden inputs_and_outputs/InSAR_dataset_test',
]

for command in commands:
    print(command)
```

## A note about safety

These notebooks do **not** automatically launch a full processing job.

That is deliberate. Stage execution can take a long time, may write outputs into the dataset folder, and may depend on external tools such as `triangle` and `snaphu`.

When you are still learning:

- start with `status`
- use `run --dry-run` before real execution
- treat verification as a read-only comparison step


## A note about `make verify`

This repository also has a `make verify` helper, but it is **not** a generic replacement for `pystamps verify --run ... --golden ...`.

Use the generic `pystamps verify` command when you want to compare arbitrary paths. The `make verify` target is only a convenience shortcut for this repository's fixed reference-path case.
